```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("C:\\Users\\sdevelop\\Desktop\\horse.jpg")

blurred = cv2.GaussianBlur(img, (5, 5), 0)
```

Gaussian Blur

```python
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Gaussian Blurred')
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Median Blur

```python
median_blurred = cv2.medianBlur(img, 5)
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Median Blurred')
plt.imshow(cv2.cvtColor(median_blurred, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Bilateral Filtering


```python
filtered = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Bilateral Filtered')
plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

Sharpening Filters

```python
blurred = cv2.GaussianBlur(img, (5, 5), 0)
mask = cv2.subtract(img, blurred)
sharpened = cv2.addWeighted(img, 1.0, mask, 1.5, 0)
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Sharpened')
plt.imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
